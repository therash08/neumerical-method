```python
def f(x):
    return x**2-2
```

```python
def bisection(xl, xu):
    if f(xl) * f(xu) > 0:
        print("invalid")
        return None

    prev_xr = 0
    it = 1

    print("it\t\t xl\t\t xu\t\t  f(xr)")
    print("-" *70)

    while True:
        xr  =  (xl + xu) / 2
        print(f"{it}\t {xl:.6f}\t {xu:.6f}\t {xr:.6f}\t{f(xr):.6f}")



        if prev_xr is not None and abs (xr-prev_xr) < 1e-6:
            break

        if f(xl) * f(xu) > 0:
            xl = xr
        else:
            xu = xr
        prev_xr  = xr
        it  += 1

        
    return xr
```
