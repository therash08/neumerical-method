```python
import math
```

```python
def f(x):
    return math.exp(-x) - x
```

```python
def df(x):
    return -math.exp(-x) - 1
```

```python
x0 = 1

while True:
    x1 = x0 - f(x0) / df(x0)
    print(f"x = {x1:.6f}")

    if round(x1, 6) == round(x0, 6):
        print("\nRoot is found:", x1)
        break

    x0 = x1
```

```python
import math
```

```python
def f(x):
    return math.exp(-x) - x
```

```python
x0 = 0
x1 = 2
es = 0.005

while True:
    if f(x0) == f(x1):
        print("Division by zero. Secant method fails.")
        break

    x = x1 - (f(x1) * (x0 - x1)) / (f(x0) - f(x1))
    print(f"x = {x:.6f}")

    ea = abs((x - x1) / x) if x != 0 else float('inf')
    print(ea, es)

    if ea < es:
        print("root is found", x)
        break

    x0 = x1
    x1 = x
```
