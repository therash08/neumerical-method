```python
def euler(f, x0, y0, h,n):
    x =x0
    y = y0
    results = [(x,y)]

    for i in range(n):
        y = y + f(x,y) * h
        x = x+ h
        results.append((x,y))
    return results
```

```python
def f(x,y):
    return -2*(x**3) + 12*(x**2)-20*x +8.5
```

```python
x0 = 0
y0 = 1
h = 0.5
n = 8


solution  = euler(f,x0,y0,h,n)

for x,y in solution:
    print(x,y)
```
