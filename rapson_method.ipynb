```python
import math
```

```python
def f(x):
    return math.exp(-x) - x
    
```

```python
def df(x):
    return -math.exp(-x) -1
```

```python
x0 = 1
while True:
    x1 = x0 - (f(x0)/df(x0))
    print(f"x = {x1:.6f}")

    if round(x1, 6) == round(x0, 6):
        print("\nroot is found", x1)
        break
    x0 = x1

```
