```python
def f(x):
  return x**2-2
```

```python
def bisection(xl, xu):
    if f(xl) * f(xu) > 0:
        print("Invalid")
        return None

    prev_xr = None
    it = 1

    print("it\t\t xl\t\t xu\t\t xr\t\t f(xr)")
    print("-" * 50)

    while True:
        xr = (xl + xu) / 2
        print(f"{it}\t{xl:.6f}\t{xu:.6f}\t{xr:.6f}\t{f(xr):.6f}")

        # stopping condition
        if prev_xr is not None and abs(xr - prev_xr) < 1e-6:
            break

        if f(xl) * f(xr) > 0:
            xl = xr
        else:
            xu = xr

        prev_xr = xr
        it += 1

    return xr
```

```python
bisection(1,2)
# print("\nRoot =", root)
```
