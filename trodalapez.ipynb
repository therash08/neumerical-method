```python
def multiple_trapezoidal(f,a,b,n):
    h = (b-a)/n
    I = f(a) + f(b)

    sum = 0
    for i in range(1,n):
        sum = sum + 2* f(a+i*h)
    I = 1/2 * h * (I + sum)
    return I
```
