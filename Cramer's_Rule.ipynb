```python
import numpy as np
```

```python
A = np.array([[2, -1, 3],
              [4, 2, -2],
              [1, 1, 1,]], dtype = float)
```

```python
b = np.array([5, 2, 3], dtype = float)
```

```python
D = np.linalg.det(A)
D
```

```python
if D == 0:
      print("No answer exists")
else:
  n = len(b)
  X = np.zeros(n)

  for i in range(n):
    Ai = np.copy(A)

    Ai[:, i] = b

    detAi = np.linalg.det(Ai)

    X[i] = detAi/D

  print(X)
```
