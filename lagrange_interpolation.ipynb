```python
def lagrange_interpolation(X, f, x_val):
    n = len(X)
    total = 0

    for i in range(n):
        term = f[i]
        for j in range(n):
            if i != j:
                term = term * (x_val - X[j]) / (X[i] - X[j])
        total += term

    return total
```

```python
X = [1, 2, 3]
f = [1, 4, 9]
```

```python
print(lagrange_interpolation(X, f, 5))
```
