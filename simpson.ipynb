```python
def f(x):
    return 0.2 + 25*x-200*(x**2)+675*(x**3)-900*(x**4)+400*(x**5)

```

```python
def simpson (f,a,b,n):
    if n%2 !=0:
        raise ValueError ("n must be even")

    h = (b-a)/n
    x = a
    result = f(a) + f(b)

    odd_sum = 0
    for i in range(1,n,2):
        odd_sum = odd_sum + f(a+i*h)
 
    even_sum = 0
    for i in range(2,n-1, 2):
        even_sum = even_sum + f(a+i*h)

    result = (1/3)*h*(result + 4*odd_sum + 2*even_sum)

    return result 

```

```python
print(simpson(f, 0, 0.8,4))
```

```python
def simpson_three_eight(f, a,b,n):
    h = (b-a)/n
    I = (3/8)* h* (f(a) + 3* f(a+h) + 3* f(a+2*h) + f(b))

    return I
```

```python
def simpson_one_third(f,a,b,n):
    h = (b-a)/n
    I = (1/3) * h* (f(a) + 4* f(a+h) + f(b))

    return I
```
